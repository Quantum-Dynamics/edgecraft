```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gate_potential = np.load("gate_potential.npy")
edge_lengths = np.load("edge_lengths.npy")
```

```python
fig, axis = plt.subplots()

axis.plot(gate_potential, edge_lengths)

axis.set_xlabel("Gate Potential ($e^2 / 4 \pi \epsilon l_0$)")
axis.set_ylabel("Edge Length ($l_0$)")

plt.show()
```
